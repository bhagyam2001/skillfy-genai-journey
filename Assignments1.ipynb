###Grade Calculator code

```python

name = input("Enter your name: ")
sub1 = int(input("Enter your marks: "))
sub2 = int(input("Enter your marks: "))
sub3 = int(input("Enter your marks: "))

Average = (sub1+sub2+sub3)/3

if(Average >= 90) :
    Grade = "A"
elif(Average >= 75 and Average < 90 ):
    Grade = "B"
elif(Average >= 60 and Average < 75 ):
    Grade = "C"
else :
    Grade = "D"

print(f'{name} got {Grade} Grade and {Average:.2f} marks in the Exams')
```

###Parking Management System Code

```python
vehicle_count = 0
total_collection = 0

while True:
    user_input = input("Enter Entry Time (or type 'exit' to stop): ")

    if user_input.lower() == "exit":
        break

    EntryTime = int(user_input)
    ExitTime = int(input("Enter Exit Time: "))

    # Validation
    if EntryTime < 6 or ExitTime > 24 or ExitTime <= EntryTime:
        print("Invalid Entry")
        continue

    ParkedDuration = ExitTime - EntryTime

    # ---- Slab Logic ----
    duration = ParkedDuration
    cost = 0

    # First 2 hours
    first_slab = min(duration, 2)
    cost += first_slab * 20

    remaining = duration - first_slab

    # Next 3 hours
    second_slab = min(remaining, 3)
    cost += second_slab * 15

    remaining -= second_slab

    # Remaining hours
    if remaining > 0:
        cost += remaining * 10

    # Vehicle counter
    vehicle_count += 1

    # 50% discount for every 5th vehicle
    if vehicle_count % 5 == 0:
        cost *= 0.5
        print("50% Discount Applied!")

    total_collection += cost

    print("Parking Duration:", ParkedDuration)
    print("Total Cost:", cost)
    print("-" * 30)

print("\nTotal Vehicles:", vehicle_count)
print("Total Revenue Collected:", total_collection)
```
